```python
import matplotlib.pyplot as plt
import matplotlib.pylab as mpl
import pandas as pd
from numpy import *
from sympy import *

plt.style.use('classic')

#Texto en latex
plt.rc('text', usetex=True)
plt.rc('axes', linewidth=2)
plt.rc('font', weight='bold')
plt.rcParams['text.latex.preamble'] = [r'\usepackage{amsmath} \boldmath']

plt.rc('xtick',labelsize=18)
plt.rc('ytick',labelsize=18)
```

```python
S, T, MH, MS, MHplus, omega1, omega2, omega, frac, sigmaSI1, sigmaSI2, LA2, LAL, LAS1, LAS2, LAS21, LAS= loadtxt("Datos_sigma2_condi_filtrados.txt", unpack= True)

M21= MH/MS
l1= abs(LAS1)
l2= abs(LAS2)

XI1= omega1 / omega
XI2= omega2 / omega
```

```python
v= 246
LA3= 2/(v**2) * (MH - MHplus)
LA21= LAL - LA3
```

```python
#------------------------------------------------------------------------------------------------------------------
#                                                     S vs T
#------------------------------------------------------------------------------------------------------------------

ax1 = plt.subplot()

x= -linspace(-0.5,0.5,1000)
y= linspace(-0.5,0.5,1000)

SS,TT= meshgrid(x,y)

eqn= 4.37421 + 718.19 * SS**2 + SS * (81.8736 - 1680.56 * TT) - 136.608 * TT + 1187.21 * TT**2
Z= [0,2.3, 6.18, 11.8]

c= mpl.cm.rainbow(linspace(0,1,6))

cp= plt.contour(SS,TT,eqn,Z, colors=c, linewidths=3)#, linestyles='dashed')
plt.clabel(cp, inline=True, fmt = '%2.2f', fontsize=12, colors= 'k')

#plt.contourf(SS,TT,eqn,Z, colors= c)

plt.scatter(S, T, color= "darkslategrey", edgecolors= "k", linewidth=0.5)
ax1 = plt.subplot()


plt.ylim(-0.1, 0.3)
plt.xlim(-0.2, 0.3)

plt.xlabel(r'$S$', fontsize=18)
plt.ylabel(r'$T$', fontsize=18)
plt.savefig('1sigma.png', dpi=500, bbox_inches='tight')

plt.show()
```
